## Attempting to Chunk a Big Document into Smaller Parts: Case Study - Qualcomm

### Document Reading

```python
import PyPDF2 as pdf
from PyPDF2 import PdfReader
import os
import pandas as pd
import re
```

```python
# Get rid of water marks and useless info
info = []
patterns = ["Confidential","80-Y8053-1 Rev.","Proprietary","Qualcomm Technologies, Inc.","MAY CONTAIN U.S. AND INTERNATIONA L EXPORT CONTROLLED INFORMATION","Wireless LAN Access Point (Driver Version 10 .4) Programmer Guide","YP  and Propr ietary –  ","5/26/2017 3:53:49 PM\nDistributed by Compex systems Pte. Ltd.\nby alex_tiong@compex.com.sg\nTo pt_myat@compex.com.sg\n","\nWireless LAN Access Point (Driver Version 10.4) Programmer Guide"]


with open("pdfs/SPF53_SPF61_programmer guide.pdf","rb") as file:
    reader = PdfReader(file)
    for j in range(0, len(reader.pages)):
        cur_string = reader.pages[j].extract_text()
        for pattern in patterns:
            cur_string = cur_string.replace(pattern,'')
        info.append("|||start of page {}".format(j+1)+cur_string + "end of page {}|||".format(j+1))
    string = ""
```

```python
# Get rid of table of contents
without_toc = info[80:]
```

### Embedding

```python
import openai
import time
import pandas as pd
import numpy as np
api_key_file = open("OPENAI_API_KEY", "r")
openai.api_key = api_key_file.read()
```

```python
from openai.embeddings_utils import get_embedding
def embed(x):
    return get_embedding(x, engine="text-embedding-ada-002")
```

```python
#Embedder (takes time)
embeddings = []
for embed_samples in without_toc:
    embeddings.append(embed(embed_samples))
    print("embedded")
```

```python
"Saved Embeddings"
np.save("SPF53_SPF61_programmer guide Embeddings Array",embeddings)
```

```python
embedding_df = pd.DataFrame({"embedding":embeddings,"text":without_toc})
```

```python
# Save CSV
embedding_df.to_csv("SPF53_SPF61_programmer guide Embeddings",index=False)
```
